```python
# =====================================================
# NeuralNet Functional Test Notebook
# =====================================================

import numpy as np
from NeuralNet import NeuralNet   # ðŸ‘ˆ Import the class from your .py file
```

```python
# XOR dataset (non-linear separable problem)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
Y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

```python
# Create a 2-2-1 architecture neural network
nn = NeuralNet([2, 2, 1])

print("Neural network initialized successfully.")
print("Architecture:", nn.n)
```

```python
# Training parameters
epochs = 5000
eta = 0.5        # Learning rate
alpha = 0.9      # Momentum

# Training loop
for epoch in range(epochs):
    total_error = 0
    for x, y in zip(X, Y):
        total_error += nn.backpropagation(x, y, eta=eta, alpha=alpha)
    
    # Display progress every 500 epochs
    if epoch % 500 == 0:
        print(f"Epoch {epoch:4d} | Total Error: {total_error:.4f}")
```

```python
print("\nðŸ§© Final outputs after training:\n")

for x in X:
    output = nn.predict(x)
    print(f"Input: {x} -> Output: {output.round(3).flatten()}")
```

```python
import matplotlib.pyplot as plt

errors = []

epochs = 5000
for epoch in range(epochs):
    total_error = 0
    for x, y in zip(X, Y):
        total_error += nn.backpropagation(x, y, eta=0.5, alpha=0.9)
    errors.append(total_error)

plt.plot(errors)
plt.title("Training Error Evolution")
plt.xlabel("Epochs")
plt.ylabel("Total Error")
plt.show()
```
