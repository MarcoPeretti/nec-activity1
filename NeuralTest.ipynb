```python
# 1 Import required libraries and the NeuralNet class

from NeuralNet import NeuralNet
import numpy as np
```

```python
# 2 Create a neural network and check its initialization


# Define network architecture: 2 input neurons, 3 hidden, 1 output
nn = NeuralNet([2, 3, 1])

# Print results

print("Neural network object created successfully.")
print("Number of layers:", nn.L)
print("Neurons per layer:", nn.n)
print("Activation function:", nn.fact)
```

```python
# Show weights and bias matrices


print(" Checking initialized weight (w) and bias (theta) matrices:")

for l in range(nn.L - 1):
    print(f"\n--- Connection {l}: Layer {l} → Layer {l+1} ---")
    print(f"Weights (w[{l}]) shape: {nn.w[l].shape}")
    print(nn.w[l])
    print(f"Biases (theta[{l}]) shape: {nn.theta[l].shape}")
    print(nn.theta[l])
```

```python
# 4 Test forward propagation skeleton


# Create a sample input vector with the same size as the input layer (2 neurons)
x = np.array([0.5, 0.2])

# Perform forward propagation
output = nn.forward(x)

print("Forward propagation result:")
print("Input:", x)
print("Network output:", output)
```

```python
# 5 Run the backpropagation skeleton 


# Define a simple target output
y_target = np.array([1.0])

# Execute the backpropagation skeleton
print("Running backpropagation skeleton...")
nn.backpropagation(x, y_target)

print("Test completed successfully!")
```
